```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "C:\\Users\\Tanisha\\Desktop\\map_object_list.csv" 
df = pd.read_csv(file_path)

# Extract relevant columns
glon = df['glon']
glat = df['glat']

# Create a scatter plot of star positions
plt.figure(figsize=(10, 6))
plt.scatter(glon, glat, s=1, alpha=0.5)
plt.xlabel('Galactic Longitude (degrees)')
plt.ylabel('Galactic Latitude (degrees)')
plt.title('Star Positions on the Sky')
plt.grid()
plt.show()
```

```python
log10_age = df['log10 age (years)']
radius = df['radius (parsecs)']

log10_age_numeric = pd.to_numeric(log10_age, errors='coerce')  # Convert non-numeric values to NaN

plt.figure(figsize=(10, 6))
plt.scatter(radius, log10_age_numeric, s=1, alpha=0.5)
plt.xlabel('Radius (parsecs)')
plt.ylabel('log10 Age (years)')
plt.title('Color-Magnitude Diagram')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

source_counts = df['Source'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
source_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Source')
plt.ylabel('Number of Stars')
plt.title('Distribution of Stars Based on Source')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(source_counts, labels=source_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proportion of Stars from Different Sources')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
