```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('C:\\Users\\Tanisha\\Desktop\\astro.csv', usecols=['DEdms', 'RAdeg', 'DEdeg', 'Vmag'])
print(data.describe())
plt.hist(data['Vmag'], bins=50, color='purple')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.title('Histogram of Star Magnitudes')
plt.show()

data['RAdeg'] = np.radians(data['RAdeg'])
data['DEdeg'] = np.radians(data['DEdeg'])

data['x'] = np.cos(data['RAdeg']) * np.cos(data['DEdeg'])
data['y'] = np.sin(data['RAdeg']) * np.cos(data['DEdeg'])
data['z'] = np.sin(data['DEdeg'])

#3D scatter plot 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['x'], data['y'], data['z'], c=data['Vmag'], cmap='magma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Hipparcos Star Catalog')
plt.show()

mean_mag = np.mean(data['Vmag'])
print('Mean magnitude:', mean_mag)

std_mag = np.std(data['Vmag'])
print('Magnitude standard deviation:', std_mag)


```
